```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_excel("https://docs.google.com/spreadsheets/d/1eUrPYL32VykYfRWIspYoBlWJk2g2CRM1e9nTX351_ds/export?format=xlsx", engine='openpyxl')
df.head()
```

```python
print("Shape of Dataset: ",df.shape)
print("\nColumns:\n", df.columns)
print("\nMissing values:\n", df.isnull().sum())
```

```python
#top run scorers
top_batsmen = df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
print(top_batsmen)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_batsmen.index, y=top_batsmen.values)
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.title('Top 10 Batsmen by Total Runs')
plt.show()
```

```python
over_runs = df.groupby('over')['batsman_runs'].mean()
plt.figure(figsize=(10, 5))
sns.lineplot(x=over_runs.index, y=over_runs.values)
plt.title('Average Runs per Over')
plt.xlabel('Over Runs')
plt.ylabel('Average Runs')
plt.show()
```

```python
powerplay = df[df["over"]<=6].groupby('batting_team')['batsman_runs'].sum()
death = df[df["over"]>=16].groupby('batting_team')['batsman_runs'].sum()
team_runs = pd.DataFrame({'Powerplay': powerplay, 'Death': death}).fillna(0)
team_runs.plot(kind = "bar",stacked=True, figsize=(10,5))
plt.title("PowerPlay vs Death Overs")
plt.xlabel("Teams")
plt.ylabel("Total Runs")
plt.show()
```

```python
df['total_runs'] = df['batsman_runs'] + df['extras']
top_batsmen = df.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
top_bowlers = df.groupby('bowler')['total_runs'].sum().sort_values(ascending=False).head(10)
matchups = df[(df["batsman"].isin(top_batsmen.index)) & (df["bowler"].isin(top_bowlers.index))]

pivot = matchups.pivot_table(
    index='batsman',
    columns='bowler',
    values='total_runs',
    aggfunc='sum',
    fill_value=0
    )
plt.figure(figsize=(12,8))
sns.heatmap(pivot, annot = True, cmap = "YlGnBu")
plt.title("Top Batsmen vs Top Bowlers")
plt.xlabel("Top Bowlers")
plt.ylabel("Top Batsmen")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```

```python
top_batsmen = df.groupby("batsman")["batsman_runs"].sum().sort_values(ascending=False).head(10).reset_index()

fig = px.bar(
    top_batsmen,
    x="batsman_runs",
    y="batsman",
    orientation="h",
    color="batsman_runs",
    color_continuous_scale="magma",
    title=" Top 10 Run Scorers in IPL"
)

fig.update_layout(
    xaxis_title="Total Runs",
    yaxis_title="Batsman",
    template="plotly_dark"
)

fig.show()
```

```python
dismissal = df["player_dismissed"].notnull()
dismissal_types = df[dismissal]["dismissal_kind"].value_counts()

fig = px.pie(
    values=dismissal_types.values,
    names=dismissal_types.index,
    title="Proportion of Dismissal Types",
    hole=0.3 # Optional: Creates a donut chart
)

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```
